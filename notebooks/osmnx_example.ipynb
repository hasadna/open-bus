````python
import folium
import osmnx as ox
````

````python
G = ox.graph_from_place('Givatayim, Israel', network_type='drive')

# There must be a built-in way to get a street name from edge, but I didn't find it
edges_dict = {(node1, node2): data for node1, node2, data in G.edges(data=True)}
````

````python
# A point we want to locate on the map
point = (32.073201, 34.808847)
````

````python
m = ox.plot_graph_folium(G)
m.add_child(folium.Marker(point))
m
````

````python
shape, node1, node2 = ox.utils.get_nearest_edge(G, point)
````
